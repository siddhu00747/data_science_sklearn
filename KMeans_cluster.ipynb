```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans  # ye hmara model hai

np.random.seed(42)
data = {'annulincome': np.random.randint(30000,100000,100),
        'spendingscore' : np.random.randint(1,100,100)}
df = pd.DataFrame(data)
df

plt.scatter(df['annulincome'],df['spendingscore'])  # x axis pr df['annulincome']   y axis pr df['spendingscore']
plt.title('coustomer data annual')
plt.xlabel('annulincome')
plt.ylabel('spendingscore')
plt.show()

x = df.values   # .values se array ki form mai data copy ho jata hai
model = KMeans(n_clusters=3 , random_state=42)
df['cluster']=model.fit_predict(x)    # ye data ko dekh kr automatic data ko group kr deta hai
df

plt.scatter(df['annulincome'],df['spendingscore'] , c=df['cluster'], cmap='rainbow')   # c color k liye use hota hai
plt.title('coustomer data annual')
plt.xlabel('annulincome')
plt.ylabel('spendingscore')
plt.show()

annu = int(input('enter your annual salary '))
spen = int(input('enter your spending score '))
user=np.array([[annu,spen]])
us = pd.DataFrame(user)

model.predict(us)

```
