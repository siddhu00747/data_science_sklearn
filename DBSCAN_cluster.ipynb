```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN  # ye hmara model hai
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data = np.random.rand(100,2)
```

```python
scaler = StandardScaler()
```

```python
df=scaler.fit_transform(data)
```

```python
model = DBSCAN(eps=0.3,min_samples=5)
```

```python
cluser = model.fit_predict(df)
```

```python
plt.scatter(df[:,0],df[:,1],c=cluser,cmap='viridis',marker='o')
plt.title('cluster')
plt.xlabel('1')
plt.ylabel('2')
```

```python
res = pd.DataFrame({'f1':df[:,0],
                    'f2':df[:,1],
                    'cluster':cluser})
print('no. of cluster',len(np.unique(cluser)))
print(res['cluster'].value_counts())
```
