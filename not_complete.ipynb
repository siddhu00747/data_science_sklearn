```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering   # ye hmara group hai
from sklearn.preprocessing import StandardScaler  
# ye group vise baat deta hai jese(different color ki balls hai tho cluster 
# ki help se hmne tho hmne ose alg alg baskets mai daal diye same color ki ek sath

np.random.seed(42)
data = pd.DataFrame({'age':np.random.randint(18,65,100),
                     'income': np.random.randint(3000,100000,100),
                     'spending': np.random.randint(1,100,100)})

# data

scaler = StandardScaler()

df = scaler.fit_transform(data)

model = AgglomerativeClustering(n_clusters=3, affinity='euclidean',linkage='ward')
model.fit(data,df)
data['cluster'] = model.fit_predict(df)

plt.scatter(data['income'],data['spending'], c=data['cluster'],cmap='viridis')
plt.xlabel('income')
plt.ylabel('spending')
plt.title('custmer')

a = int(input('enter your age'))
income = int(input('enter you income'))
spen = int(input('enter your spendig'))

user = pd.DataFrame({'age':[a],
                     'income':[income],
                     'spending':[spen]})
us = scaler.transform(user)
model.fit_predict([[us]])

```
